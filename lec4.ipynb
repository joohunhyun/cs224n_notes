#### Formal Definition of Dependency Grammar

Formally : Dependency syntax assumes that syntactic structure consists of relations between **lexical
items**, normally binary asymmetric relations (“arrows”) called dependencies

<p align="center">
    <img src="img/j/l4/5.png" alt="Word Vectors" width="500"/>
</p>

- arrows are **typed** with the name of grammatical relations
- dependencies form a tree that is:
  - connected
  - acyclic
  - has a single head
- Tree banks are good because :
  - reusability of labor
  - broad coverage

#### Dependency Conditioning : Rule of Thumb

- Bilexical affinities
  - The dependency [discussion → issues] is plausible
- Dependency distance
  -  Most dependencies are between nearby words
- Intervening material
  - Dependencies rarely span intervening verbs or punctuation
- Valency of heads
  - How many dependents on which side are usual for a head?

### Dependency Parsing : Method

- constraints
  - only one word is a dependent of ROOT
  - no cycles
- projective or non-projective
  - TLDR: can arrows cross each other(non-projective)? or not(projective)?
  - We usually stick to projective parsers

#### Projectivity

**Projective parse** : There are no crossing dependency arcs when the
words are laid out in their linear order, with all arcs above the words
